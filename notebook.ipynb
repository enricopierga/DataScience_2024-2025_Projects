```python
# Filtra i match di Federer (sia come Player_1 che come Player_2)
federer_matches = df[(df['Player_1'] == 'Federer R.') | (df['Player_2'] == 'Federer R.')].copy()

# Estrai l'anno dalla colonna Date
federer_matches['Year'] = pd.to_datetime(federer_matches['Date']).dt.year

# Calcola vittorie, sconfitte e percentuale vittorie per anno
federer_stats = []
for year in federer_matches['Year'].unique():
    year_matches = federer_matches[federer_matches['Year'] == year]
    
    # Conta vittorie quando Federer Ã¨ Player_1 o Player_2
    wins = len(year_matches[
        ((year_matches['Player_1'] == 'Federer R.') & (year_matches['Winner'] == 'Federer R.')) |
        ((year_matches['Player_2'] == 'Federer R.') & (year_matches['Winner'] == 'Federer R.'))
    ])
    
    # Il totale delle partite meno le vittorie sono le sconfitte
    total_matches = len(year_matches)
    losses = total_matches - wins
    
    # Calcola la percentuale di vittorie
    win_percentage = (wins / total_matches) * 100
    
    federer_stats.append({
        'Year': year, 
        'Vittorie': wins, 
        'Sconfitte': losses,
        'Percentuale_Vittorie': win_percentage
    })

# Converti in DataFrame
federer_stats_df = pd.DataFrame(federer_stats)

# Crea la figura
plt.figure(figsize=(18, 12))
plt.style.use('seaborn')

# Plot delle vittorie e sconfitte (barre)
plt.bar(federer_stats_df['Year'], federer_stats_df['Vittorie'], 
        color='#27ae60', label='Vittorie', alpha=0.7, width=0.8)
plt.bar(federer_stats_df['Year'], federer_stats_df['Sconfitte'], 
        bottom=federer_stats_df['Vittorie'], color='#e74c3c', 
        label='Sconfitte')

# Personalizzazione del grafico
plt.title('Statistiche di Roger Federer per Anno', pad=20, fontsize=24, fontweight='bold')
plt.xlabel('Anno', fontsize=18, fontweight='bold')
plt.ylabel('Numero di Partite', fontsize=18, fontweight='bold')

# Griglia e stile
plt.grid(True, linestyle='--', alpha=0.3)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Rotazione delle etichette sull'asse x
plt.xticks(rotation=45, fontsize=14)
plt.yticks(fontsize=14)

# Aggiungi la legenda con font piÃ¹ grande
plt.legend(loc='upper left', fontsize=16)

# Aggiungi le etichette delle percentuali sopra ogni barra
for i, row in federer_stats_df.iterrows():
    total_height = row['Vittorie'] + row['Sconfitte']
    plt.annotate(f'%W: {row["Percentuale_Vittorie"]:.1f}%',
                xy=(row['Year'], total_height),
                xytext=(0, 5), textcoords='offset points',
                ha='center', va='bottom',
                fontsize=9,
                bbox=dict(boxstyle='round,pad=0.5', fc='white', ec='gray', alpha=0.7))

plt.tight_layout()
plt.show()
```

```python
# Funzione per calcolare le statistiche di vittorie per un giocatore
def calcola_statistiche_vittorie(df, player_name):
    player_matches = df[(df['Player_1'] == player_name) | (df['Player_2'] == player_name)].copy()
    player_matches['Year'] = pd.to_datetime(player_matches['Date']).dt.year
    
    player_stats = []
    for year in player_matches['Year'].unique():
        year_matches = player_matches[player_matches['Year'] == year]
        wins = len(year_matches[year_matches['Winner'] == player_name])
        total_matches = len(year_matches)
        win_percentage = (wins / total_matches) * 100 if total_matches > 0 else 0
        
        player_stats.append({
            'Year': year, 
            'Percentuale_Vittorie': win_percentage
        })
    
    return pd.DataFrame(player_stats)

# Calcola le statistiche per ciascun giocatore
federer_stats_df = calcola_statistiche_vittorie(df, 'Federer R.')
nadal_stats_df = calcola_statistiche_vittorie(df, 'Nadal R.')
djokovic_stats_df = calcola_statistiche_vittorie(df, 'Djokovic N.')

# Crea il grafico
plt.figure(figsize=(16, 9))
plt.style.use('seaborn-darkgrid')

# Plot delle percentuali di vittorie
sns.lineplot(data=federer_stats_df, x='Year', y='Percentuale_Vittorie', marker='o', label='Federer', color='#1f77b4', linewidth=2.5)
sns.lineplot(data=nadal_stats_df, x='Year', y='Percentuale_Vittorie', marker='o', label='Nadal', color='#ff7f0e', linewidth=2.5)
sns.lineplot(data=djokovic_stats_df, x='Year', y='Percentuale_Vittorie', marker='o', label='Djokovic', color='#2ca02c', linewidth=2.5)

# Personalizzazione del grafico
plt.title('Andamento Percentuale di Vittorie per Anno', pad=20, fontsize=22, fontweight='bold')
plt.xlabel('Anno', fontsize=16, fontweight='bold')
plt.ylabel('Percentuale di Vittorie (%)', fontsize=16, fontweight='bold')

# Griglia e stile
plt.grid(True, linestyle='--', alpha=0.5)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Rotazione delle etichette sull'asse x
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

# Aggiungi la legenda
plt.legend(loc='upper left', fontsize=14)

# Annotazioni per i punti di massimo e minimo
for stats_df, color in zip([federer_stats_df, nadal_stats_df, djokovic_stats_df], ['#1f77b4', '#ff7f0e', '#2ca02c']):
    max_year = stats_df.loc[stats_df['Percentuale_Vittorie'].idxmax()]['Year']
    max_value = stats_df['Percentuale_Vittorie'].max()
    plt.annotate(f'Max: {max_value:.1f}%', xy=(max_year, max_value), xytext=(5, 5), textcoords='offset points', ha='center', color=color, fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Carica il file CSV
df = pd.read_csv('atp_tennis.csv')

# Carica il dataset mondiale
world = gpd.read_file('110m_cultural/ne_110m_admin_0_countries.shp')

# Calcola il numero di tornei unici per ciascun paese
tornei_unici_per_paese = df.groupby('Country')['Tournament'].nunique().reset_index()
tornei_unici_per_paese.columns = ['Country', 'Numero_Tornei']

# Unisci i dati usando 'ADMIN'
world = world.merge(tornei_unici_per_paese, how='left', left_on=['ADMIN'], right_on=['Country'])
world['Numero_Tornei'] = world['Numero_Tornei'].fillna(0)

# Assicurati che il massimo valore di Numero_Tornei sia maggiore di tutti i valori nei bin
max_tornei = world['Numero_Tornei'].max()
bins = [0, 1, 5, 10, 20, max_tornei + 1]  # Aggiungi 1 per garantire che il massimo sia incluso

# Crea bins personalizzati per la legenda
world['Tornei_Categoria'] = pd.cut(world['Numero_Tornei'], 
                                   bins=bins, 
                                   labels=[f'{bins[i]}-{bins[i+1]}' for i in range(len(bins)-1)])

# Imposta lo stile della mappa
fig, ax = plt.subplots(1, 1, figsize=(20, 10))

# Crea la mappa con stile personalizzato
world.boundary.plot(ax=ax, linewidth=1, color='black')  # Aggiungi contorni ai paesi
world.plot(column='Numero_Tornei',
           ax=ax,
           legend=True,
           legend_kwds={'label': 'Numero di Tornei ATP Unici per Paese',
                        'orientation': 'horizontal',
                        'shrink': 0.8},
           missing_kwds={'color': 'lightgrey'},
           cmap='viridis')  # Cambia la scala di colori a 'viridis'

# Aggiungi titolo e personalizzazioni
plt.title('Distribuzione Mondiale dei Tornei ATP Unici\n(Conteggio Tornei per Paese)', 
          fontsize=16, 
          pad=20)

# Rimuovi assi
ax.axis('off')

plt.tight_layout()
plt.show()
```
