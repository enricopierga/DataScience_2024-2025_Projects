```python
# Importazione delle librerie necessarie
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Caricamento del dataset
# Sostituisci 'path_to_dataset.csv' con il percorso al tuo file CSV contenente i dati
try:
    df = pd.read_csv('Datasets/atp_tennis.csv')
except FileNotFoundError:
    print("Errore: Il file specificato non è stato trovato. Verifica il percorso.")
```

Analisi della Distribuzione dei Campi Odd_1 e Odd_2

```python
# Configurazione estetica dei grafici
sns.set_theme(style="whitegrid", palette="pastel")

# Rimozione dei valori nulli o gestione
df = df.dropna(subset=['Odd_1', 'Odd_2'])  # Rimuove le righe con valori nulli in Odd_1 o Odd_2

# Creazione degli istogrammi
plt.figure(figsize=(14, 6))

# Istogramma per Odd_1
plt.subplot(1, 2, 1)
sns.histplot(df['Odd_1'], bins=20, kde=True, color="blue", stat="density", edgecolor="black")
plt.title('Distribuzione di Odd_1')
plt.xlabel('Odd_1')
plt.ylabel('Densità')
plt.xlim(0, 20)

# Istogramma per Odd_2
plt.subplot(1, 2, 2)
sns.histplot(df['Odd_2'], bins=20, kde=True, color="orange", stat="density", edgecolor="black")
plt.title('Distribuzione di Odd_2')
plt.xlabel('Odd_2')
plt.ylabel('Densità')
plt.xlim(0, 20)

# Mostra i grafici
plt.tight_layout()
plt.show()
```

Codice per la Distribuzione di Rank_1 e Rank_2

```python
# Pulisci i dati rimuovendo eventuali valori nulli
df = df.dropna(subset=['Rank_1', 'Rank_2'])

# Esplora la distribuzione dei ranking
plt.figure(figsize=(14, 6))

# Istogramma per Rank_1 (Giocatore 1)
plt.subplot(1, 2, 1)
sns.histplot(df['Rank_1'], bins=20, kde=True, color="blue", edgecolor="black")
plt.title('Distribuzione di Rank_1 (Giocatore 1)')
plt.xlabel('Rank_1')
plt.ylabel('Frequenza')
plt.xlim(0, 1000)

# Istogramma per Rank_2 (Giocatore 2)
plt.subplot(1, 2, 2)
sns.histplot(df['Rank_2'], bins=20, kde=True, color="orange", edgecolor="black")
plt.title('Distribuzione di Rank_2 (Giocatore 2)')
plt.xlabel('Rank_2')
plt.ylabel('Frequenza')
plt.xlim(0, 1000)

# Mostra i grafici
plt.tight_layout()
plt.show()
```

Distribuzione delle Vittorie per Giocatore

```python
# Conta le vittorie per ciascun giocatore
winner_counts = df['Winner'].value_counts()

# Seleziona i primi 30 giocatori più vincenti
top_20_winners = winner_counts.head(30)

# Esplora la distribuzione delle vittorie
plt.figure(figsize=(10, 8))

# Istogramma con i giocatori sull'asse Y e il numero di vittorie sull'asse X
sns.barplot(x=top_20_winners.values, y=top_20_winners.index, palette='viridis', edgecolor='black')

# Titolo e etichette
plt.title('Top 20 Giocatori Più Vincenti')
plt.xlabel('Numero di Vittorie')
plt.ylabel('Giocatori')
plt.xticks(ticks=range(0, int(top_20_winners.values.max()) + 100, 100))

# Mostra il grafico
plt.tight_layout()
plt.show()
```

 Istogrammi di Rank_1, Rank_2, Pts_1 e Pts_2

```python
# Esplora la distribuzione di Rank_1, Rank_2, Pts_1, Pts_2
plt.figure(figsize=(14, 10))

# Istogramma per Rank_1 (Rank del Giocatore 1)
plt.subplot(2, 2, 1)
sns.histplot(df['Rank_1'], bins=20, kde=True, color="blue", edgecolor="black")
plt.title('Distribuzione di Rank_1 (Giocatore 1)')
plt.xlabel('Rank_1')
plt.ylabel('Frequenza')

# Istogramma per Rank_2 (Rank del Giocatore 2)
plt.subplot(2, 2, 2)
sns.histplot(df['Rank_2'], bins=20, kde=True, color="orange", edgecolor="black")
plt.title('Distribuzione di Rank_2 (Giocatore 2)')
plt.xlabel('Rank_2')
plt.ylabel('Frequenza')

# Istogramma per Pts_1 (Punteggio del Giocatore 1)
plt.subplot(2, 2, 3)
sns.histplot(df['Pts_1'], bins=20, kde=True, color="green", edgecolor="black")
plt.title('Distribuzione di Pts_1 (Giocatore 1)')
plt.xlabel('Pts_1')
plt.ylabel('Frequenza')

# Istogramma per Pts_2 (Punteggio del Giocatore 2)
plt.subplot(2, 2, 4)
sns.histplot(df['Pts_2'], bins=20, kde=True, color="red", edgecolor="black")
plt.title('Distribuzione di Pts_2 (Giocatore 2)')
plt.xlabel('Pts_2')
plt.ylabel('Frequenza')

# Mostra i grafici
plt.tight_layout()
plt.show()
```

Grafico a barre per il numero di partite giocate su ciascuna superficie o in ciascun round.

```python
# Grafico a barre per il numero di partite giocate su ciascuna superficie
plt.figure(figsize=(14, 6))

# Grafico a barre per la superficie
plt.subplot(1, 2, 1)
sns.countplot(data=df, x='Surface', palette='viridis', edgecolor='black')
plt.title('Numero di Partite per Superficie')
plt.xlabel('Superficie')
plt.ylabel('Numero di Partite')
plt.xticks(rotation=45)  # Ruota le etichette delle superfici per una lettura migliore

# Grafico a barre per il numero di partite giocate in ciascun round
plt.subplot(1, 2, 2)
sns.countplot(data=df, x='Round', palette='viridis', edgecolor='black')
plt.title('Numero di Partite per Round')
plt.xlabel('Round')
plt.ylabel('Numero di Partite')
plt.xticks(rotation=45)  # Ruota le etichette dei round per una lettura migliore

# Mostra i grafici
plt.tight_layout()
plt.show()
```

matrice di correlazione tra le variabili Rank_1, Rank_2, Pts_1, e Pts_2

```python
#Ha più senso Pts e Partite Vinte (da calcolare) per anno




# Seleziona solo le colonne di interesse per la correlazione
columns_of_interest = ['Rank_1', 'Rank_2', 'Pts_1', 'Pts_2']

# Calcola la matrice di correlazione
correlation_matrix = df[columns_of_interest].corr()

# Visualizza la matrice di correlazione come una heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Titolo e visualizzazione
plt.title('Matrice di Correlazione tra Rank_1, Rank_2, Pts_1 e Pts_2')
plt.tight_layout()
plt.show()
```

```python
# Conta le vittorie per ciascun tennista e seleziona i 30 più vincenti
top_30_winners = df['Winner'].value_counts().head(30).index

# Filtra il dataset per includere solo i 30 tennisti più vincenti
df_top_30 = df[df['Winner'].isin(top_30_winners)]

# Creazione dei grafici
plt.figure(figsize=(16, 12))

# Grafico per i tennisti più vincenti su "Clay"
plt.subplot(2, 2, 1)
clay_winners = df_top_30[df_top_30['Surface'] == 'Clay']['Winner'].value_counts()
sns.barplot(x=clay_winners.values, y=clay_winners.index, palette='viridis', edgecolor='black')
plt.title('Vincitori su Clay')
plt.xlabel('Numero di Vittorie')
plt.ylabel('Tennisti')
plt.xticks(rotation=45)

# Grafico per i tennisti più vincenti su "Hard"
plt.subplot(2, 2, 2)
hard_winners = df_top_30[df_top_30['Surface'] == 'Hard']['Winner'].value_counts()
sns.barplot(x=hard_winners.values, y=hard_winners.index, palette='viridis', edgecolor='black')
plt.title('Vincitori su Hard')
plt.xlabel('Numero di Vittorie')
plt.ylabel('Tennisti')
plt.xticks(rotation=45)

# Grafico per i tennisti più vincenti su "Grass"
plt.subplot(2, 2, 3)
grass_winners = df_top_30[df_top_30['Surface'] == 'Grass']['Winner'].value_counts()
sns.barplot(x=grass_winners.values, y=grass_winners.index, palette='viridis', edgecolor='black')
plt.title('Vincitori su Grass')
plt.xlabel('Numero di Vittorie')
plt.ylabel('Tennisti')
plt.xticks(rotation=45)

# Grafico per i tennisti più vincenti su "Carpet" (se presente)
plt.subplot(2, 2, 4)
carpet_winners = df_top_30[df_top_30['Surface'] == 'Carpet']['Winner'].value_counts()
sns.barplot(x=carpet_winners.values, y=carpet_winners.index, palette='viridis', edgecolor='black')
plt.title('Vincitori su Carpet')
plt.xlabel('Numero di Vittorie')
plt.ylabel('Tennisti')
plt.xticks(rotation=45)

# Layout e visualizzazione
plt.tight_layout()
plt.show()
```
